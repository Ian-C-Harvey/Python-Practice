```python
tenYearDF = df[df['Date'] <= '2012-01-01']
```

```python
def Daily_R(Open, Close): 
    Daily_return = Open - Close
    return round(Daily_return,5)

df['DR'] = tenYearDF[
    ['Open','Close']].apply(
    lambda df: Daily_R(
        df['Open'],df['Close']), axis=1)
```
