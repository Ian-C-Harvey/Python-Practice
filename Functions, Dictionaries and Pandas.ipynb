```python
#For iteration after creating the dictionary
# the range information can be found by typing df.index and using stop + 1 as the end range.
index_column = range(0,12564)

# Each of these are considered a series and have to be converted to a list for dictionary comprehension.
Date = df.Date.values.tolist()
Open = df.Open.values.tolist()
High = df.High.values.tolist()
Low = df.Low.values.tolist()
Close = df.Close.values.tolist()
Volume = df.Volume.values.tolist()

# Dictionary comprehension for practice. This code creates one big list with keys.
NewDict = {
    index_column: [Date, Open, High, Low, Close, Volume]
    for index_column, Date, Open, High, Low, Close, Volume in zip(
        index_column, Date, Open, High, Low, Close, Volume)
        }

# Outputting part of the dictionary to prevent Juypter Notebook from crashing
y = 1
while y < 15:
    print(NewDict[y][::])
    y +=1
print('---')

# Zip creates a tuple and dict makes the first value into the dictionary key.
# Decided not to add the index column so I coulde use the date as the keys and critiera. This makes iteration hard.
NewDict = dict((zip(df.Date, 
                    zip(df.Open, df.Close, 
                         zip(df.High, df.Low,df.AdjClose)))))


# Using dates as the key for a dicitonary with a lisit containing all values and adding those values using a for loop 
DatePreciseDict = {}
for keys, values in NewDict.items():
    if keys < '2012-01-01':
        continue
    else: 
        DatePreciseDict[keys] = values

x = 0
while x < 1:
    print({DatePreciseDict[keys][::]})
    x +=1
# Increasing the size of x would cause the same value to be printed multiple times.
```

```python
def Summarize_Values(listvalues):
    ''' 
    Takes in a list of numerical values and provides summary information.
    Args:
        listvalues (float): List of values for the summarization

    Returns:
        Std: Standard deviation
        Var: Variance
        Average: Sum of values / length
        Unique_values: Creates a set from the list.
    '''    
    Average = sum(listvalues)/len(listvalues)
    Arraylist = np.array(listvalues)
    Var = np.var(Arraylist)
    Std = np.sqrt(np.var(Arraylist))
    Unique_values = set(listvalues)
    Largerthanaverage = Arraylist > Average
    TrueCountVal = np.count_nonzero(Largerthanaverage)
    
    print(f'{TrueCountVal} value(s) are greater than the average')
    print(Unique_values)
    print('Average', 'Variance', 'Standard Deviation')
    return [Average, Var, Std, Unique_values]

DictOpenList = []
DictCloseList = []

for x,y in DatePreciseDict.items():
    DictOpenList.append(round((y[0]),0))

for x,y in DatePreciseDict.items():
    DictCloseList.append(round((y[1]),0))

Average, Var, Std, Unique_Values = Summarize_Values(DictOpenList)
print(Average, Var, Std)
print("-----")
Average1, Var1, Std1 , Unique_Values1 = Summarize_Values(DictCloseList)
print(Average, Var, Std)
print("-----")

print('Values present in both sets')
try:
    Difference = Unique_Values.intersection(Unique_Values1)
    print(Difference)
except NameError:
    pass

# The two list are different rounding to 0 made them very similar and the output more readable.
```
